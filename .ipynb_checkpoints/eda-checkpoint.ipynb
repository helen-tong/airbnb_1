```python
def compact_names(data):
    name_list=[]
    for line in data:
        name_list.append(line)
    return name_list
```

```python
top95 = compact_names(top95.name)
top95
```

```python
def word_dict(x):
    d ={}
    for line in x:
        for word in line.lower().split():
            if word in d:
                d[word] += 1
            else:
                d[word] = 1
    return d
```

```python
name_dict_top95 = word_dict(top95)
```

```python
def pets(row):
    if isinstance(row['house_rules'], str):
        if 'no pet' in row['house_rules']:
            row['pets'] = False
        else:
            row['pets'] = True
    else:
        row['pets'] = True
    return row
```
